#ondas en un estanque

simulacion de una onda en un estanque por una sola perturbacion

```python
from pylab import plot, imshow, show
from numpy import loadtxt, array, zeros, sqrt, pi, sin
from pylab import xlabel, ylabel, imshow, plot,show

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = 50. # cm
y1 = 50. # cm

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for xx in range( len(x) ):
    for yy in range( len(y) ):
        r = sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        outfile.write("%0.1f " % (zeta0*sin(k*r)))
    
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
imshow(data, origin="lower", extent=[0, 100, 0, 100])
xlabel("Distancia cm")
ylabel("Distancia cm")
show()
```

#Interferencia entre ondas

simulacion de dos ondas generadas en un estanque con una separacion de 20 cm en la posicion x

```python
from pylab import plot, imshow, show
from numpy import loadtxt, array, zeros, sqrt, pi, sin
from pylab import xlabel, ylabel, imshow, plot,show

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = float(input("ingrese el valor posicion en x..:"))
y1 = float(input("ingrese el valor posicion en y..:"))
x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for yy in range( len(y) ):
    for xx in range( len(x) ):
        r1 = sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        r2 = sqrt( ((x1+20)-x[xx])*((x1+20)-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        zeta1=zeta0*sin(k*r1)
        zeta2=zeta0*sin(k*r2)
        outfile.write("%0.1f " % (zeta1))
        outfile.write("%0.1f " % (zeta2))
    
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
imshow(data, origin="lower", extent=[0, 100, 0, 100])
xlabel("propagacion en x (cm)")
ylabel("propagacion en y (cm)")
show()
```

#simulacion del patron de interferencia por la suma de las dos ondas

```python
from pylab import plot, imshow, show
from numpy import loadtxt, array, zeros, sqrt, pi, sin
from pylab import xlabel, ylabel, imshow, plot,show

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for yy in range( len(y) ):
    for xx in range( len(x) ):
        r1 = sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        r2 = sqrt( ((x1+20)-x[xx])*((x1+20)-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        zeta1=zeta0*sin(k*r1)
        zeta2=zeta0*sin(k*r2)
        outfile.write("%0.1f " % (zeta1+zeta2))
        
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
imshow(data, origin="lower", extent=[0, 100, 0, 100])
xlabel("propagacion en x (cm)")
ylabel("propagacion en y (cm)")
show()
```
